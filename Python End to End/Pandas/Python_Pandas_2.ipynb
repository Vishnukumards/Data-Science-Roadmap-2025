```python
import pandas as pd
import numpy as np

# Example DataFrame with a timestamp column
data = {
    'timestamp': pd.date_range('2025-05-01 00:00', periods=6, freq='5T'),  # 5-minute frequency
    'value': [1, 2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)
# Set the 'timestamp' column as the index
df.set_index('timestamp', inplace=True)

# Apply a rolling window of 10 minutes ('10T') and calculate the sum over each window
df['rolling_sum'] = df['value'].rolling('10T').sum()

# Print the DataFrame to see the results
print(df)
```

```python
df.rolling(window=1).max()
```

```python
df.rolling(window=1).mean()
```

```python
df.rolling(window=1).sum()
```
