```python
good_data = BeautifulSoup(response.content, "html.parser")
```

```python
img_list = good_data.find_all("img")
```

```python
import requests
import os

image_data_mongo = []
for i in img_list:
    image_url = i["src"]
    image_data = requests.get(image_url).content  # Download image content
    my_dict = {"index": image_url, "image": image_data}
    image_data_mongo.append(my_dict)
    
    # Saving image to disk
    with open(os.path.join(save_dir, f"{data}_{img_list.index(i)}.jpg"), "wb") as f:
        f.write(image_data)  # Use .write() to save the image data
```
