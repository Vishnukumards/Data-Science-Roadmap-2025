```python
# Import

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
```

```python
plt.plot(x,y)                
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
```

```python
x = np.linspace(-3,3,30)
y = x**2
plt.plot(x,y,'r')
plt.show()
```

```python
import math

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title("Sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.plot(x,y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
# scatter is used to know the relationship of Data
plt.scatter(x,y)  
```

```python
plt.figure(figsize= (6,2))
plt.scatter(x,y, c='#223BDC',alpha = .9 , marker = "x")
plt.xlabel(" X axis Data")
plt.ylabel(" Y axis Data")
plt.title("X and Y data")
plt.grid()
```

```python
x = np.linspace(1,10 , 100)
y = np.sin(x)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y , '--r')   # colour with marker  '--r'
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sin wave graph")
plt.grid()
```

```python
Stream =[ 2009,	2010, 2011]
Arts= [50,	70,	75]
```

```python
plt.figure(figsize=(5,3))
plt.bar( Stream , Arts)
plt.xlabel("Stream Data")
plt.ylabel(" Arts Data")
plt.title(" DATA REPRESENTATION")
```

```python
plt.figure(figsize=(5,3))
plt.bar( Stream , Arts)
plt.xlabel("Stream Data")
plt.ylabel(" Arts Data")
plt.title(" DATA REPRESENTATION")
plt.grid()
```

```python
plt.figure(figsize=(5,3))  # Horizontal Bar Chart"
plt.barh( Stream , Arts)
plt.xlabel("Stream Data")
plt.ylabel(" Arts Data")
plt.title(" DATA REPRESENTATION")
```
