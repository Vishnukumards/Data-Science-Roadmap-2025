```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
# Generate a random n-class classification problem.
# make_classification helps in creating Random datasets
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
# Binary Classification
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
# NOw for the above data i will create Support Vector Classifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc =SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
# Prediction
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print("---Classification Report---")
print(classification_report(y_pred,y_test))
print("---Confusion Matrix---")
print(confusion_matrix(y_test,y_pred))
print("--Accuracy Score----")
print(accuracy_score(y_test,y_pred))
```

```python
# Hyperparameter tuning WIth SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['linear']
}
```

```python
gs_svc = GridSearchCV(SVC(),param_grid=parameter,refit=True,cv=5,verbose=3)
```

```python
gs_svc.fit(X_train,y_train)
```

```python
gs_svc.best_params_
```

```python
# Prediction after Hyperparameter Tuning
y_pred = gs_svc.predict(X_test)
print("---Classification Report---")
print(classification_report(y_test,y_pred))
print("---Confusion Matrix---")
print(confusion_matrix(y_test,y_pred))
print("--Accuracy Score----")
print(accuracy_score(y_test,y_pred))
```
