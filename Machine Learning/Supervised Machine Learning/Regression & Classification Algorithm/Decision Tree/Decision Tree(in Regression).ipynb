#### 🔧 Hyperparameter Tuning in Machine Learning

## 📌 What is Hyperparameter Tuning?

Hyperparameter tuning is the process of **selecting the best set of parameters** for a machine learning model to ensure it learns **effectively and accurately**.

These parameters are not learned from the data — they are **set before training** and control how the model learns.

---


### What is Cross-Validation?
Cross-validation is a technique to evaluate how well a machine learning model will generalize to unseen data.

It does not split the data by model type (e.g., 50% to Decision Tree and 50% to SVM). Instead, it splits the data into multiple folds, trains on some folds, and tests on the remaining — repeating this across all folds.

**So hyperparameter select the parameters and cross validation tells which model with which parameter is working best**

## Parameters in decision tree 
| Parameter                  | Type           | Description                                                                                              |
| -------------------------- | -------------- | -------------------------------------------------------------------------------------------------------- |
| `max_depth`                | int            | Maximum depth of the tree. Limits how many levels the tree can grow. Prevents overfitting.               |
| `min_samples_split`        | int/float      | Minimum number of samples required to split an internal node. Higher values stop tree from growing deep. |
| `min_samples_leaf`         | int/float      | Minimum number of samples required to be at a leaf node. Forces the model to keep leaves bigger.         |
| `max_leaf_nodes`           | int            | Maximum number of terminal (leaf) nodes in the tree. Limits complexity directly.                         |
| `min_weight_fraction_leaf` | float          | The minimum weighted fraction of the sum total of weights required to be at a leaf node.                 |
| `max_features`             | int/float/auto | Limits the number of features to consider at each split.                                                 |
| `ccp_alpha`                | float          | Post-pruning parameter (cost-complexity pruning). Higher values prune more aggressively.                 |


```python
## NOW
from sklearn.tree import DecisionTreeRegressor

parameter = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5, 6, 7, 8, 10, 11, 12],
    'max_features': ['sqrt', 'log2']
}
regressor =DecisionTreeRegressor()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV
regressorcv =GridSearchCV(regressor,param_grid=parameter,cv=2,scoring='neg_mean_squared_error')
```

```python
regressorcv.fit(X_train,y_train)
```

```python
regressorcv.best_params_
```

```python
dt_clf_regressor = DecisionTreeRegressor(criterion='absolute_error', max_depth = 11,max_features= 'log2',splitter='best')
```

```python
dt_clf_regressor.fit(X_train,y_train)
```

```python
y_pred= dt_clf_regressor.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_pred,y_test)
```
