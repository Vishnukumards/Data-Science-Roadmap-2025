# Purity Split Check :-

# 1.Entropy (Root node calculate)

```python
#### P(+) is Probability of Positive Category
#### P(-) is Probability of Negative Category
```

## Entropy is the randomness in the information being processed

It measures the purity of the split. It is hard to draw conclusions from the information when the entropy increases. It ranges between 0 to 1. 1 means that it is a completely impure subset.

# 2.Gini Impurity

```python
#### Gini Impurity = 1-[(P+)^2+(P-)^2]
```

Gini index is also type of criterion that helps us to calculate information gain. It measures the impurity of the node and is calculated for binary values only.

# 3.Gain Ratio

The gain ratio is the modification of information gain. It takes into account the number and size of branches when choosing an attribute. It takes intrinsic information into account.

# GR(S,A) = Gain( S,A)/ IntI(S,A)

# 1.Information Gain

```python
#### H(s) is Root node
#### H(Sv) is Entropy of Category 
#### |Sv| is Category sum
#### |S| is Total number of sample sum
```

Information gain is a decrease in entropy. Decision trees make use of information gain and entropy to determine which feature to split into nodes to get closer to predicting the target and also to determine when to stop splitting.

## Which to use Entropy or Gini Impurity ?

```python
When we have small set of features-> Use Entropy [log Formula]
When we have large set of features -> Use Gini Impurity [Simple Maths]
```

# Understand the Target and Features
Target: PlayTennis

Features: Outlook, Temperature, Humidity, Windy

Calculate the Entropy of the Target :

#### Formula is above :

=(-5/8)log2(5/8)-(3/8)log2(3/8)

Entropy=0.954


```python
Subset Split by Outlook:
Sunny: 3 examples → [No, No, No]
Overcast: 2 examples → [Yes, Yes]
Rain: 3 examples → [Yes, Yes, No]
```

| Subset       | Label Counts (Yes/No) | Formula                                            | Values Substituted                                                                         | Entropy (Rounded) |
| ------------ | --------------------- | -------------------------------------------------- | ------------------------------------------------------------------------------------------ | ----------------- |
| **Sunny**    | Yes = 0, No = 3       | $-p_{Yes} \log_2(p_{Yes}) - p_{No} \log_2(p_{No})$ | $-0 \cdot \log_2(0) - 1 \cdot \log_2(1)$ = $-0 - 0$                                        | **0.00**          |
| **Overcast** | Yes = 2, No = 0       | $-p_{Yes} \log_2(p_{Yes}) - p_{No} \log_2(p_{No})$ | $-1 \cdot \log_2(1) - 0 \cdot \log_2(0)$ = $-0 - 0$                                        | **0.00**          |
| **Rain**     | Yes = 2, No = 1       | $-p_{Yes} \log_2(p_{Yes}) - p_{No} \log_2(p_{No})$ | $-\frac{2}{3} \log_2\left(\frac{2}{3}\right) - \frac{1}{3} \log_2\left(\frac{1}{3}\right)$ | **0.918**         |


```python
Calculate ig for one by one:
IG,Outlook = Entropy(D)−((3/8)Entropy(Sunny)+(2/8)Entropy(Overcast)+(3/8)Entropy(Rain))
```

```python
IG,Outlook = 0.954-((3/8)0+(2/8)0+(3/8)0.918)
```

```python
Where,
Entropy of Dataset (D) = 0.954
```

```python
(3/8)*0+(2/8)*0+(3/8)*0.918  # Entropy After Split on Outlook
```

```python
 0.954-((3/8)*0+(2/8)*0+(3/8)*0.918) #Information Gain
```
