# Data Interpolation

Data interpolation is the process of estimating unknown values with the data set based on known values

1. Linear Interpolation
2. Cublic Interpolation with Scipy
3. Polynomial Interpolation

# Linear Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(x,y)
```

```python
# Interpolating x and y by creating  and adding new linspace data to the x and y by interp function using numpy
x_new = np.linspace(1,5,10)
y_interp = np.interp(x_new,x,y)
```

```python
x_new
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
## Cubic Interpolation with scipy
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d  # interp1d is one dim
```

```python
f = interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_inter = f(x_new)
```

```python
y_inter
```

```python
plt.title("Before Interpolation")
plt.scatter(x,y)
```

```python
plt.title("After Interpolation")
plt.scatter(x_new,y_inter)
```

```python
# Polynomial Interpolation
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
# interpolate the data using polynomial interpolation
p = np.polyfit(x,y,2)
```

```python
x_new= np.linspace(1,5,10) # create new x values
y_interp = np.polyval(p,x_new) # interpolate y values
```

```python
plt.title("Before Polynomial Interpolation")
plt.scatter(x,y)
```

```python
plt.title("After Polynomial Interpolation")
plt.scatter(x_new,y_interp)
```
