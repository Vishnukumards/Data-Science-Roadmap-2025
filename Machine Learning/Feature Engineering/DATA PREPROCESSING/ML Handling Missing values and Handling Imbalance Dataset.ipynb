#### DOWNSAMPLING: It means feature which has higher data will be Downsampled to lower data

```python
##Example : A [DATA]    B [DATA]
#             800         200
    
## HERE BY USING  DOWNSAMPLING WE UPDATE THE DATA OF B[DATA] - TO 1000
    
```

```python
np.random.seed(123)

sample = 1000
data_ = 0.8
a_value = int(sample*data_)
b_value = sample - a_value 
```

```python
A_value =pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size= a_value),
    'feature_2':np.random.normal(loc=0,scale=1,size= a_value),
    'target':[0]*a_value
})
B_value = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size= b_value),
    'feature_2':np.random.normal(loc=0,scale=1,size= b_value),
    'target':[1]*b_value
})
```

```python
df=pd.concat([A_value,B_value]).reset_index(drop=True)
```

```python
print(df['target'].value_counts())
```

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample
```

```python
df_majority_downsampled = resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsample = pd.concat([df_minority,df_majority_downsampled])
```

```python
df_downsample['target'].value_counts()
```

# 3.SMOTE(Synthetic Minority Oversampling Technique)

#### SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

# Data Interpolation

#### Data interpolation is the process of estimating unknown values within a dataset based on the known values.

```python
import numpy as np
```

```python
x= np.array([11,12,13,14,15])
```

```python
y= np.array([21,22,23,24,25])
```

```python
# LINEAR INTERPOLATION
```

```python
plt.scatter(x,y)
```

```python
# Creating New X values using the above data 
```

```python
xp=np.linspace(11,15,10) # Sample points
yp=np.interp(xp,x,y)  # here x an y be our data One-dimensional linear interpolation for monotonically increasing sample points. 
```

```python
xp
```

```python
yp
```

```python
plt.scatter(xp,yp)
```

```python
# CUBIC INTERPOLATION (Scipy)
```

```python
xc = np.array([1,2,3,4,5,6])
yc = np.array([1,8,23,48,100,155])
```

```python
from scipy.interpolate import interp1d
```

```python
new_data= interp1d(xc,yc,kind='cubic')
```

```python
xc_new = np.linspace(1,5,10)
yc_new = new_data(xc_new)       #yc-interpted data from xc_new using cubic  
```

```python
# OLD DATA

plt.scatter(xc,yc)
```

```python
# NEW DATA

plt.scatter(xc_new,yc_new)
```

```python
# POLYNOMIAL INTERPOLATION
```

```python
xply = np.array([1,2,3,4,5])
yply = np.array([1,4,9,16,25])
```

```python
new_data_poly= np.polyfit(xply,yply,3)
```

```python
xply_new = np.linspace(1,5,10)
yply_new = np.polyval(new_data_poly,xply_new)
```

```python
# OLD DATA
```

```python
plt.scatter(xply,yply)
```

```python
# NEW DATA
```

```python
plt.scatter(xply_new,yply_new)
```

#### A Bagging classifier with additional balancing. This implementation of Bagging is similar to the scikit-learn implementation. It includes an additional step to balance the training set at fit time using a given sampler.

```python

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
bbc = BalancedBaggingClassifier(random_state=42)
bbc.fit(X_train, y_train)

BalancedBaggingClassifier(...)
y_pred = bbc.predict(X_test)
print(confusion_matrix(y_test, y_pred))
```
