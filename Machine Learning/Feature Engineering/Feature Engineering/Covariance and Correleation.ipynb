### What Is Covariance?

### Covariance is a statistical term that refers to a systematic relationship between two random variables in which a change in the other reflects a change in one variable.

```python
# Select only numeric columns
numeric_df = df[['Year', 'Spending_USD', 'Life_Expectancy']]

# Covariance matrix
print(numeric_df.cov())

# Covariance heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(numeric_df.cov(), annot=True, cmap='coolwarm')
plt.title("Covariance Matrix Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Original data
data = {
    "Area": ["IN Banglore", "OutSide Banglore", "In Village", "IN Town"],
    "HousePrice": [12000, 7000, 2000, 5000]
}

# Step 2: Convert to DataFrame
df = pd.DataFrame(data)

# Step 3: Map Area to numeric values
area_mapping = {
    "IN Banglore": 4,
    "OutSide Banglore": 3,
    "IN Town": 2,
    "In Village": 1
}
df['AreaNumeric'] = df['Area'].map(area_mapping)

# Step 4: Compute covariance and correlation
cov_matrix = df[['AreaNumeric', 'HousePrice']].cov()
corr_matrix = df[['AreaNumeric', 'HousePrice']].corr()

print("Covariance Matrix:\n", cov_matrix)
print("\nCorrelation Matrix:\n", corr_matrix)

# Step 5: Plot heatmaps
plt.figure(figsize=(12, 5))

# Covariance
plt.subplot(1, 2, 1)
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm')
plt.title("Covariance Matrix")

# Correlation
plt.subplot(1, 2, 2)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")

plt.tight_layout()
plt.show()
```

| Concept                | Covariance                    | Correlation                                |
| ---------------------- | ----------------------------- | ------------------------------------------ |
| **What it tells you**  | Direction of relationship     | Direction **and strength** of relationship |
| **Scale**              | Unbounded (can be very large) | Always between **-1 and +1**               |
| **Affected by units?** | Yes (e.g., \$ vs kg matters)  | ❌ No — it's **unitless**                   |
| **Hard to compare?**   | Yes                           | ❌ No — easy to compare across variables    |

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
Y = np.array([10, 20, 30, 40, 50])

cov_matrix = np.cov(X, Y)
print("Positive Covariance:\n", cov_matrix)
```

```python
np.corrcoef(X,Y)    # Pearson product-moment correlation coefficients.
```

```python
# primary function in NumPy to calculate correlation.
```

```python
np.correlate(X,Y)  # Cross-correlation of two 1-dimensional sequences.
```

```python
from scipy.stats import pearsonr
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50])
r, p = pearsonr(x, y)
print("Pearson correlation:", r)
print("P-value:", p)
```

```python
from scipy.stats import spearmanr

r, p = spearmanr(x, y)
print("spearman correlation:", r)
print("P-value:", p)
```

```python
from scipy.stats import kendalltau

r, p = kendalltau(x, y)
print("kendalltau correlation:", r)
print("P-value:", p)
```

# P-Value

### P value tells you the probability of Your data where if P value is less than 0.05 So the relationship between the variables is significant if the P value is greater than 0.05 then I can conclude that there is no relationship between the variables they occur due to random chances.

| Method      | Correlation Coefficient | P-value                    |
| ----------- | ----------------------- | -------------------------- |
| Pearson     | 1.0                     | 0.0                        |
| Spearman    | ≈ 1.0                   | **1.404e-24** (very small) |
| Kendall Tau | ≈ 1.0                   | **0.0166**                 |


| P-value Range | Interpretation                         | Conclusion                        |
| ------------- | -------------------------------------- | --------------------------------- |
| `< 0.01`      | Very strong evidence against null (H₀) | **Highly significant**            |
| `0.01 – 0.05` | Moderate evidence                      | **Statistically significant**     |
| `0.05 – 0.1`  | Weak evidence                          | May be **suggestive**, not strong |
| `> 0.1`       | No significant evidence                | **Likely random/noise**           |
