# 1. Standardization
Formula:



```python
import numpy as np
```

```python
# We need to find Mean and Standard deviation
```

```python
# Mean 
mean = np.mean(data['total_bill'])
std = np.std(data['total_bill'])

# Now we found
print(" Mean    &    Standard deviation") 
print(mean, " &  " ,std)  
```

```python
Standardization_bill = []

for i in list(data['total_bill']):
    z_score = (i - mean)/std
    Standardization_bill.append(z_score)
```

```python
sns.histplot(data['total_bill'])
```

```python
sns.histplot(Standardization_bill)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(data[['total_bill']])
```

```python
scaler.transform(data[['total_bill']])
```

```python
import pandas as pd
pd.DataFrame(scaler.fit_transform(data[['total_bill', 'tip']]), columns= ['total_bill','tip'])
```

```python
# So here scaling for new data
scaler.transform([[13,4]])
```

# 2. Normalization (min Max Scaler)
Formula:



```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
import numpy as np

xmin = np.min(df['passengers'])
xmax = np.max(df['passengers'])
```

```python
maxx_min = []
for i in list(df['passengers']):
    min_max = (i - xmin)/(xmax - xmin)
    maxx_min.append(min_max)
```

```python
maxx_min[2]
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
pd.DataFrame(minmax.fit_transform(df[['distance','fare','tip']]),columns=['distance','fare','tip'])
```

```python
# So here scaling for new data
minmax.transform([[1.14,7.16, 7.15]])
```
