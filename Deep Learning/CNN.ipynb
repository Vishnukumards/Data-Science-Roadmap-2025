# 2.Alexnet Architecture

The Alexnet has eight layers with learnable parameters the model consists of five layers with a combination of Max Pooling followed by three fully connected layers and they use relu activation in each of these layer except the output layer so we already know about Lenet architecture above.


- The architecture employs overlapping pooling layers to reduce spatial dimensions while retaining the spatial relationships among neighbouring features.

- Activation function: AlexNet uses the ReLU activation function and dropout regularization, which enhance the model's ability to capture non-linear relationships within the data.

The key features of AlexNet are as follows:-
- AlexNet was created to be more computationally efficient than earlier CNN topologies. It introduced parallel computing by utilising two GPUs during training.
- AlexNet is a relatively shallow network compared to GoogleNet. It has eight layers, which makes it simpler to train and less prone to overfitting on smaller datasets.
- In 2012, AlexNet produced ground-breaking results in the ImageNet Large Scale Visual Recognition Challenge (ILSVRC). It outperformed prior CNN architectures greatly and set the path for the rebirth of deep learning in computer vision.
- Several architectural improvements were introduced by AlexNet, including the use of rectified linear units (ReLU) as activation functions, overlapping pooling, and dropout regularisation. These strategies aided in the improvement of performance and generalisation

## In order to solve the problem of the vanishing/exploding gradient, this architecture introduced the concept called Residual Blocks. In this network, we use a technique called skip connections.

The approach behind this network is instead of layers learning the underlying mapping, we allow the network to fit the residual mapping. So, instead of say H(x), initial mapping, let the network fit, 

    F(x) := H(x) - x which gives H(x) := F(x) + x. 

In Backward Propagation while we back propagate the loss need to travel back direclty and it travel in skip connection path.This keeps over gradient alive (Vanishing gradient Problem).

# 5.Inception Net (GoogLeNet) 

The Inception architecture combines 1x1, 3x3, and 5x5 convolutional filters in parallel to extract features from the input image at different scales. The output of each filter is then concatenated and passed on to the next layer, where the process is repeated.

Why Inception?

Traditional CNNs (like VGG) stack layers sequentially. But what if we dont know the best kernel size to capture patterns?
Inception solves this by doing multiple convolutions in parallel and concatenating the outputs.

### Basic Building Block: The Inception Module

 Instead of choosing one filter size, Inception uses:
- 11 Convolution
- 33 Convolution
- 55 Convolution
- 33 Max Pooling

Each of these runs in parallel on the same input, and their outputs are concatenated.

```python
               Input
                 
 
                                
          11 conv               
          33 conv               
          55 conv               
        33 max pool             
                                
  Concatenate Outputs 
                 
              Output
```

1x1 Convolution  The Secret Ingredient
- Used for dimensionality reduction before expensive 33 or 55 convolutions.
- Reduces the computational cost drastically.

Example:
Instead of feeding 256 channels into a 5x5 conv, use a 1x1 conv to reduce them to 64 first.

Auxiliary Classifiers (During Training)

To help with gradient flow in deep networks, GoogLeNet uses auxiliary softmax classifiers connected at intermediate layers. Theyre only used during training.

### Parameter Efficiency

Despite being deeper and more powerful, Inception Net has fewer parameters than VGGNet due to:

- Dimensionality reduction via 11 convs
- Carefully balanced depth vs width

### Advantages of Inception
- Captures features at multiple scales
- Efficient in terms of computation and memory
- Easy to extend with deeper layers (Inception v2/v3)
