### Weight Initialization

Weight initialization is a crucial step in training neural networks. It involves setting the initial values of the weights in the network before the training process begins. The goal of proper weight initialization is to prevent issues like vanishing or exploding gradients during training, which can hinder the network's ability to learn effectively.

**Why is it important?**

*   **Avoiding Vanishing/Exploding Gradients:** If weights are initialized too small, gradients can shrink during backpropagation (vanishing gradients), making it difficult for the network to learn. If weights are initialized too large, gradients can grow uncontrollably (exploding gradients), leading to unstable training.
*   **Speeding up Convergence:** Good initialization can help the optimization algorithm converge faster to a good solution.
*   **Breaking Symmetry:** If all weights are initialized to the same value, all neurons in a layer will learn the same things, making the network less powerful. Random initialization breaks this symmetry.

### Types of Weight Initialization

There are several popular weight initialization techniques:

1.  **Zero Initialization:** Setting all weights to zero. This is generally a bad idea as it leads to the symmetry problem mentioned above.
2.  **Random Initialization:** Initializing weights with small random values. This is better than zero initialization, but the scale of the random values is important.
3.  **Xavier/Glorot Initialization:** This method is designed to keep the variance of the activations the same across layers. It's suitable for layers with activation functions like sigmoid and tanh.
4.  **He Initialization:** This method is similar to Xavier initialization but is designed for layers with the ReLU activation function. It accounts for the fact that ReLU outputs are zero for negative inputs.



# Example of random initialization

```python
# Import Libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
weights_random = np.random.randn(10, 5) * 0.01 # Small random values
print("Random Initialization:")
print(weights_random)
```

### Example: Xavier/Glorot Initialization

Xavier initialization scales the random values based on the number of input and output neurons.

```python
# Example of Xavier/Glorot initialization
input_dim = 10
output_dim = 5
limit = np.sqrt(6 / (input_dim + output_dim))
weights_xavier = np.random.uniform(-limit, limit, size=(input_dim, output_dim))
print("\nXavier/Glorot Initialization:")
print(weights_xavier)
```

```python
# Example of He initialization
input_dim = 10
weights_he = np.random.randn(input_dim, output_dim) * np.sqrt(2 / input_dim)
print("\nHe Initialization:")
print(weights_he)
```

### Best Initialization for Which Activation Function?

*   **Sigmoid and Tanh:** Xavier/Glorot initialization is generally recommended for these activation functions.
*   **ReLU and its variants (Leaky ReLU, ELU):** He initialization is typically the best choice for these activation functions.

# Zero Initializer

```python
# Data Set - Mnist Data
# Mnist Data is the pixcel (28x28) image of 0 to 9 numbers written image in different ways
```

```python
mnist = tf.keras.datasets.mnist
```

```python
mnist
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
X_train[0]
```

```python
X_test[0]
```

```python
# Create and validate data set from the full training data
# scale the data between 0 to 1 by dividing it by 225, as its an unsigned data between 0-255 range
x_valid = X_train[:5000]/255.0  # last five thousand images
x_train = X_train[5000:]/255.0   # first five thousand images
y_valid = y_train[:5000]
y_train = y_train[5000:]

# scale the test set as well
x_test = X_test/255.0
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(x_train[0],annot=True,cmap="binary")
plt.show()
```

```python
# Flatten mean convert 2D to 1D (our pixcel image is 28x28 => to 784 (1D))
# First hidden layer with 300 neurons And Activation function relu- Introduce nonlinearity. Why 300? First hidden layer with 300 neurons.
# Second hidden layer with 100 neurons.  learned from the previous layer, enabling deeper understanding.
# Output layer with 10 neurons, each representing one digit class (0 to 9). Softmax convert raw scores into probabilitiessumming tp 1
```

```python
from keras.layers import Dense, Flatten,Dropout
```

```python
model_clf = Sequential()
# Sequential - groups a linear stack of layers
model_clf.add(Flatten(input_shape=[28,28],name="inputLayer"))
model_clf.add(Dense(units=64, activation='relu',kernel_initializer=tf.keras.initializers.Zeros()))
model_clf.add(Dropout(0.2))
model_clf.add(Dense(units=32, activation='relu',kernel_initializer=tf.keras.initializers.Zeros()))
model_clf.add(Dropout(0.1))
model_clf.add(Dense(units=16, activation='relu',kernel_initializer=tf.keras.initializers.Zeros()))
model_clf.add(Dropout(0.2))
model_clf.add(Dense(units=10,activation='softmax'))
```

```python
model_clf.summary()
```

```python
Loss_function = "sparse_categorical_crossentropy"
# FOR LOSS FUNCTION use : tf.losses.SparseCategoricalCrossentropy()

Optimizer = "adam"
# FOR Optimizer use: tf.keras.optimizers.Adam()

Metrics = ["accuracy"]

# Hyperparameter Tuning
model_clf.compile(loss=Loss_function,optimizer=Optimizer,metrics=Metrics)
```

```python
# In neural network at iteration we can check the error and we can learn from it
EPOCHS = 5
VALIDATION_SET = (x_valid,y_valid)
history = model_clf.fit(x_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET,batch_size=32)
```

```python
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.show()
```

```python
# In Zero Initialize It will not allow model to learn Properly so we can see the Validation and Accuracy Drop
```

# Xavier normal initializer

### Glorot normal initializer, also called Xavier normal initializer.

```python
model_clf = Sequential()
# Sequential - groups a linear stack of layers
model_clf.add(Flatten(input_shape=[28,28],name="inputLayer"))
model_clf.add(Dense(units=64, activation='relu',kernel_initializer=tf.keras.initializers.GlorotNormal()))
model_clf.add(Dropout(0.2))
model_clf.add(Dense(units=32, activation='relu',kernel_initializer=tf.keras.initializers.GlorotNormal()))
model_clf.add(Dropout(0.1))
model_clf.add(Dense(units=16, activation='relu',kernel_initializer=tf.keras.initializers.GlorotNormal()))
model_clf.add(Dropout(0.2))
model_clf.add(Dense(units=10,activation='softmax'))
```

```python

# Hyperparameter Tuning
model_clf.compile(loss=Loss_function,optimizer=Optimizer,metrics=Metrics)
```

```python
# In neural network at iteration we can check the error and we can learn from it
EPOCHS = 5
VALIDATION_SET = (x_valid,y_valid)
history = model_clf.fit(x_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET,batch_size=32)
```

```python
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.show()
```

```python
# Observation from Diagram: The model shows high training accuracy and low loss, but the rising validation loss indicates overfitting. While validation accuracy remains stable, the model is not generalizing well to unseen data.
```

# He Normalization

```python
model_clf = Sequential()
# Sequential - groups a linear stack of layers
model_clf.add(Flatten(input_shape=[28,28],name="inputLayer"))
model_clf.add(Dense(units=64, activation='relu',kernel_initializer=tf.keras.initializers.HeNormal(seed=None)))
model_clf.add(Dropout(0.2))
model_clf.add(Dense(units=32, activation='relu',kernel_initializer=tf.keras.initializers.HeNormal(seed=None)))
model_clf.add(Dropout(0.1))
model_clf.add(Dense(units=16, activation='relu',kernel_initializer=tf.keras.initializers.HeNormal(seed=None)))
model_clf.add(Dropout(0.2))
model_clf.add(Dense(units=10,activation='softmax'))
```

```python

# Hyperparameter Tuning
model_clf.compile(loss=Loss_function,optimizer=Optimizer,metrics=Metrics)
```

```python
# In neural network at iteration we can check the error and we can learn from it
EPOCHS = 5
VALIDATION_SET = (x_valid,y_valid)
history = model_clf.fit(x_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET,batch_size=32)
```

```python
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.show()
```
