```python
# Data Set - Mnist Data
# Mnist Data is the pixcel (28x28) image of 0 to 9 numbers written image in different ways
```

```python
mnist = tf.keras.datasets.mnist
```

```python
mnist
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
# Create and validate data set from the full training data
# scale the data between 0 to 1 by dividing it by 225, as its an unsigned data between 0-255 range
x_valid = X_train[:5000]/255.0  # last five thousand images
x_train = X_train[5000:]/255.0   # first five thousand images
y_valid = y_train[:5000]
y_train = y_train[5000:]

# scale the test set as well
x_test = X_test/255.0
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# Data Show
plt.imshow(x_train[0],cmap="binary")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(x_train[0],annot=True,cmap="binary")
plt.show()
```

```python
# Creating layers of ANN + Batch Normalize
Layers = [tf.keras.layers.Flatten(input_shape=[28,28],name="inputLayer"),
          tf.keras.layers.Dense(300,activation="relu",name="hiddenLayer1"),
          tf.keras.layers.BatchNormalization(), # scale 300 data and find gamma * X + Beta
          tf.keras.layers.Dense(100,activation="relu",name="hiddenLayer2"),
          tf.keras.layers.BatchNormalization(),
          tf.keras.layers.Dense(10,activation="softmax",name="outputLayer")]
```

```python
# Flatten mean convert 2D to 1D (our pixcel image is 28x28 => to 784 (1D))
# First hidden layer with 300 neurons And Activation function relu- Introduce nonlinearity. Why 300? First hidden layer with 300 neurons.
# Second hidden layer with 100 neurons.  learned from the previous layer, enabling deeper understanding.
# Output layer with 10 neurons, each representing one digit class (0 to 9). Softmax convert raw scores into probabilitiessumming tp 1
```

```python
model_clf = tf.keras.models.Sequential(Layers)
# Sequential - groups a linear stack of layers
```

```python
model_clf.layers
```

```python
784*300 + 300
# 235,200 (weights) + 300 (biases) = 235,500 trainable parameters
```

```python
300*100+100
# 300 * 100 + 100 = 30,100 parameters
```

```python
100*10+10
```

| Layer      | Weights             | Biases | Total       |
| ---------- | ------------------- | ------ | ----------- |
| Dense(300) | 784 × 300 = 235,200 | 300    | **235,500** |
| Dense(100) | 300 × 100 = 30,000  | 100    | **30,100**  |
| Dense(10)  | 100 × 10 = 1,000    | 10     | **1,010**   |
| **Total**  |                     |        | **266,610** |


```python
model_clf.summary()
```

```python
# Now above we also have Batch Normalization
# After 2 dense layer apply batch normalization layer "enough"
```

```python
# access the Input layers
Input = model_clf.layers[0]
Input.name
```

```python
# access the layers
hidden1 = model_clf.layers[1]
hidden1.name
```

```python
# access the layers hidden layer Index changes because hidden layer may be from two to many
hidden2 = model_clf.layers[2]
hidden2.name
```

```python
# access the layers Output layes - It is also changes based on the hidden layer
output = model_clf.layers[3]
output.name
```

```python
# We can also get per Layers  - Hidden layer one
weights,biases =hidden1.get_weights()
print(weights.shape)
print(biases.shape)
```

```python
# We can also get per Layers  - Hidden layer Two
gamma, beta, moving_mean, moving_variance = hidden2.get_weights()
print(f"Gamma shape: {gamma.shape}")
print(f"Beta shape: {beta.shape}")
print(f"Moving mean shape: {moving_mean.shape}")
print(f"Moving variance shape: {moving_variance.shape}")
```

```python
Loss_function = "sparse_categorical_crossentropy"
# FOR LOSS FUNCTION use : tf.losses.SparseCategoricalCrossentropy()

Optimizer = "adam"
# FOR Optimizer use: tf.keras.optimizers.Adam()

Metrics = ["accuracy"]

# Hyperparameter Tuning
model_clf.compile(loss=Loss_function,optimizer=Optimizer,metrics=Metrics)
```

```python
# Batch_size is 32 , 32 batches selected for each Epoch 1
```

```python
# In neural network at iteration we can check the error and we can learn from it
EPOCHS = 30
VALIDATION_SET = (x_valid,y_valid)
history = model_clf.fit(x_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET,batch_size=32)
```

```python
pd.DataFrame(history.history)
```

```python
# We can see that the accuracy of from the first epochs to the 30th epochs it has been increased
```

```python
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.show()
```

```python
# Observation from Diagram: The model shows high training accuracy and low loss, but the rising validation loss indicates overfitting. While validation accuracy remains stable, the model is not generalizing well to unseen data.
```

```python
# We can evaluate our model
```

```python
model_clf.evaluate(x_test,y_test)
```
